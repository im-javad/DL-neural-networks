# Softmax

## Intro

Softmax is an activation function used in the output layer of multi-class classification models.
It converts raw scores (logits) into probabilities where:
- All values become between 0 and 1
- The sum of all outputs = 1
- Each value represents the probability of a class
<br/><br/>



## Formula

$$
\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$

## Example (Before vs After Softmax)

Raw model outputs (logits): [2.1, 0.5, -1.2]

After applying Softmax: [0.81, 0.16, 0.03]

***Softmax transforms arbitrary numbers into a probability distribution.***

## Why do we use Softmax?

- It turns model outputs into probabilities
- Allows comparison between classes
- Works perfectly with CrossEntropy loss

## Key Notes

- Use softmax in the output layer for multi-class problems
- Use `categorical_crossentropy` when labels are one-hot
- Use `sparse_categorical_crossentropy` when labels are class indices
- Softmax ensures the output is a valid probability distribution
