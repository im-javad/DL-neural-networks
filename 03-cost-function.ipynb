# Intro

- A criterion for measuring how bad the model is
    - Badness? the difference between the model's output and the expected output
- We use the following symbols and variables
    - : representative of expected output (true value)
    - Y: result estimate of network output
<br/> <br/>



- The simple cost function for classification formula:
    - C = ( - y)^2 / n
- Larger errors are more prominent due to the square power
    - 0.1^2 = 0.01 | 0.4^2 = 0.16

- But simple cost function slows down the training process
- So use of ***cross entropy***
    - C = -i i log(yi)
- This func let us to faster training
- The greater the diffrence, the faster the neuron can learn
- And also, because of the help of log(yi), it makes errors more prominent
