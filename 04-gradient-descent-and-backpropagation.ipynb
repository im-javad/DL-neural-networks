# Gradient Descent

## Intro

- An optimization algorithm for find min of a function
- To find a local minimum of a function using this algorithm, steps proportional to the negative gradient of the function at the current location will be taken

## How it work?

- The main element in gradient descent is the derivative
- We proceed with the help of the chart's slope and efforts to reduce the slope
- We can go from the right peak to the valley instead of from the left peak to the valley
<br/> <br/>



## Why gradient decent and not derivative decent?

- We use of gradient decent because loss function in ML are multidimensional.
- A simple derivative works only in 1D, but a gradient provides the direction and magnitude of steepest descent in N-dimentional parameter space.
<br/><br/>



## MLP (multy layer perceptron) 

- Only for the last layer, in addition to the neurons own output, we also have the expected output(labels)
- We are looking the method that we calculate this weights gradient compared to the final error function
    - We havn't no labels (or output values) for neurons prior to the final layer, so we can't calculate their derivatives because we don't have the    correct values for the neurons in these layers to say that i compute the correct values for them and compute a cost function to update the neurons.
    - So we use of ***chain rule***
        - This law says that if we have functions written with respect to each other, you can compute the final derivative by multiplying their partial derivatives.
<br/>



## Backpropagation

- When we give an image to the network, we go from input to output (or the term feedforward)
- But during optimization or training, we calculate a loss function or cost function at the end, and the last layer is updated first, then the previous layer, and the one before that, until we reach the initial layer
- Backpropagation means we calculate the effects of each weight relative to the final error using these derivatives
<br/> <br/>



#### **tip**: Error backpropagation us used to calculate the error contibution of each neuron agter processing a batch of data

#### **tip:** This method heavily relies on the chain rule to find derivatives, to go back through the network and calculate these errors

## Important Tips

That's why we say that if networks are single-layered, they use gradient descent, and if they are multi-layered, they like to use gradient descent, but sine they don't have derivatives, they use backpropagation to calculate the derivatives, then apply then with gradient descent and perform the update.
