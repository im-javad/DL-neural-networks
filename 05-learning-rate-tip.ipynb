# What number should I put at all?

- The closer we get to the optimal point, the smaller the update steps and the gradient derivative slope become.
- Of course, sometimes algorithms manually reduce the LR to help when you're getting close to the solution, so you proceed with more caution and don't overshoot, or in simple terms, don't pass a place 30 meters away at 100 km/h.
- To find a suitable learning rate, we definitely have to experiment through trial and error, and the correct learning rate is never constant, and if it were, our job would be completely easy.
