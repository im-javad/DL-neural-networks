# Dropout

## Intro

Dropout is a regularization method that reduces overfitting by randomly deactivating a subset of neurons during training. This prevents strong co-adaptation between neurons and forces the model to learn more general features. It also acts as an implicit ensemble of many smaller networks, improving overall generalization performance.

In summary: by randomly killing neurons in the layers. <br/>
e.g during training of a network we use only half of those parameters, and during testing the entire network works for us to make predictions 

## Why Does Dropout Reduce Overfitting?
- Prevents Co-adaptation of Neurons
    - Since different neurons are removed randomly in each training step, the network cannot rely on specific pathways. It is forced to learn more general and robust features.
- Acts Like Model Averaging
    - Every training iteration with dropout effectively creates a slightly different sub-model.
The final network behaves like an ensemble of these many sub-models, which improves stability and generalization.
- Introduces Controlled Noise
    - Randomly dropping neurons works like injecting noise into the model.
This noise prevents the network from memorizing training data and helps it perform better on unseen data.

## How Dropout Works?
- During Training:
    - A fixed ratio of neurons (e.g., 0.3 for 30%) are randomly ignored â€” their outputs are set to zero.
- During Testing:
    - All neurons are active, but their outputs are scaled appropriately to maintain the same expected value as during training.

## Practical Notes
- Highly effective in Dense (Fully Connected) layers.
- Typical dropout rate: 0.2 to 0.5.
- Very high dropout rates (e.g., 0.7) may cause underfitting.
