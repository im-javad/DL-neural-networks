# ReLU (Rectified Linear Unit)

- Formula is => max(0, z) (non-linear).
- Actually, it says to set the negative parts to zero (0) and keep the positive parts as they are (1).
- In neural networks, we will mostly deal with positive features.
- With the sigmoid and tan(h) functions we couldn't have more than two hidden layers, and this function solved the problem.
- ReLU is important because it avoids the vanishing-gradient problem, making deep networks learn much faster.
- It is extremely simple and computationally efficient, and it produces sparse activations that improve stability and reduce overfitting.
- Its piecewise-linear behavior also helps neural networks model complex nonlinear functions effectively.
