# What's perceptron (an neuron)

the biological neuron: 
<br/> <br/> 



- Artificial neurons have both inputs and outputs
- This simple model is called a perceptron
- Inputs values are features
- Inputs are multiplied by the weights
- The weights are randomly initialized (E.g -1, 0.5, ...)
- A activation function is applied to the results
- We have many selections for activation function
    - The simplest activation function is:
        - if total was a positive => 1
        - otherwise => 0
- We can add a bias when all inputs are zero and the function output becomes zero for every weight to solve the problem
